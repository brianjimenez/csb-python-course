### Example 1: Plot one function

````python
%matplotlib inline

# First import is needed to setup the matplotlib lib
import matplotlib
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-2*np.pi, 2*np.pi, 0.1)

plt.plot(x, np.cos(x))

# Adding some axis labels
plt.xlabel('x')
plt.ylabel('cos(x)')

# Show the plot
plt.show()
````

### Example 2: Plot two functions

````python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-2*np.pi, 2*np.pi, 0.1)

plt.plot(x, np.cos(x), 'r', x, np.sin(x), 'g')

# Equivalent:
# plt.plot(x, np.cos(x), 'r')
# plt.plot(x, np.sin(x), 'g')

plt.xlabel('x')
plt.ylabel('cos(x)')

plt.show()
````

### Example 3: Gaussian

````python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab

mu, sigma = 50, 7

x = mu + sigma * np.random.randn(10000)

n, bins, patches = plt.hist(x, 30, density=1, facecolor='dodgerblue', alpha=0.5)

plt.xlabel('x')
plt.ylabel('Prob(x)')

# We can place text!
plt.text(20, .025, r'$\mu=50,\ \sigma=7$')

plt.axis([10, 90, 0, 0.07])

plt.grid(True)

plt.show()
````

### Example 5: 3D function

````python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

step = 0.2

x = np.arange(-2*np.pi, 2*np.pi, step)
y = np.arange(-2*np.pi, 2*np.pi, step)
x, y = np.meshgrid(x, y)
f = np.cos(x) - np.sin(y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(x, y, f, rstride=1, cstride=1, cmap='coolwarm',
                       linewidth=0, antialiased=False)

plt.show()
````
